###  Age Analysis:

```python
# Age Distribution of Dataset
x = Data.Age
plt.hist(x)
plt.xlabel("Age")
plt.ylabel("Number of players")
plt.title("Players age distribution", fontsize=16)
plt.show()
print('Taking a look at this visualization, we can see that the mean and median age of players is between 20 - 25')
print('Median Age = ', Data.Age.median())
print('Mean Age = ', Data.Age.mean())
```

```python
#Youngest Players
youngest_Players = Data.loc[Data['Age'] == Data.Age.min(),['Name','Nationality','Age','Overall','Club']]
youngest_Players.set_index('Name', inplace=True)
print(youngest_Players)
```

```python
#Look at Age distribution in 5 best footballing nations
selected_nations = ("Spain", "England", "France", "Italy", "Germany")
# Then we assign the value from the data set
nations_age = Data[Data["Nationality"].isin(selected_nations) & Data["Age"] ]

#Then we make a violin plot
sns.violinplot(x="Nationality", y="Age", data=nations_age)
sns.set_context("paper")
```

```python
#Look at Age Distribution in some elite clubs
selected_clubs = ("FC Barcelona", "Juventus", "Paris Saint-Germain", "Manchester City", "FC Bayern München")
# Then we assign the value from the data set
club_age = Data[Data["Club"].isin(selected_clubs) & Data["Age"] ]

#Then we make a violin plot
sns.violinplot(x="Club", y="Age", data=club_age)
sns.set_context("paper")
```

### Wage Analysis:

Convert wages and values into float type instead of object type.(€85.5M will be 85500000, €6.1K will be 6100, €0 will be 0.)

```python
def convert_value(v):
    if('M' in str(v)):
        return float(v[1:-1]) * 1e6
    elif('K' in str(v)):
        return float(v[1:-1]) * 1e3  
    elif('€0' in str(v)):
        return float(v[1:])
    else:
        return v
```

```python
Data['Value'] = Data['Value'].apply(convert_value)
Data['Wage'] = Data['Wage'].apply(convert_value)
Data['Release Clause'] = Data['Release Clause'].apply(convert_value)
```

```python
#Collect list of all players with wages expect those with "0" wages who are free agents
Data = Data.dropna(subset = ['Club'])
```

```python
#Top 5 most payed players
most_Payed = Data.sort_values("Wage",ascending=False)[['Name','Wage','Age','Overall','Club']].set_index('Name').head(5)
print("Most Payed Players\n",most_Payed)

#Top 5 least payed players 
least_Payed = Data.sort_values("Wage",ascending=True)[['Name','Wage','Age','Overall','Club']].set_index('Name').head(5)
print("Least Payed Players\n",least_Payed)
```

```python
#5 Most Expensive Squad (Perfom a groupby Club and add player wages of each club)
most_exp_squad = Data.groupby('Club')['Wage'].sum().nlargest()
print(most_exp_squad)
```
